```python
import pandas as pd
```

```python
olympics = pd.read_excel('olympics-data.xlsx')

olympics.sample(10)
```

```python
olympics[olympics['born_country'] == 'CRO']
```

```python
# filtering with regex (anything can be passed on the contain method)
olympics[olympics['NOC'].str.contains('CRO|col', case=False, na=False)]
```

```python
# We can concatenate any number of conditions. str is a string method that allows us use any python string based method
olympics[olympics['born_country'].str.contains('BRA', na = False, case=False) & (olympics['height_cm'] > 170)]
```
